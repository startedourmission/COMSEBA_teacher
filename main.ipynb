```python
import sheet_manager

SPREADSHEET_ID = ''
TODAY = '2025-05-17'  
data_all = sheet_manager.get_sheet_data(spreadsheet_id=SPREADSHEET_ID, range_name = '동춘!A3:Z1000')
data = {}
for k, v in data_all.items():
    data[k] = [v[11], v[12]]
data
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
from webdriver_manager.chrome import ChromeDriverManager
import time

# 브라우저 옵션 설정
chrome_options = Options()
# chrome_options.add_argument("--headless")  # 헤드리스 모드 (화면에 표시 안 함)
chrome_options.add_argument("--start-maximized")  # 창 최대화

# 웹드라이버 설정
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
driver.get("https://w3.a-b.kr/login.aspx?ReturnUrl=%2ff1.aspx")

try:
    # 웹사이트 접속
    
    
    # 로그인
    username = driver.find_element(By.ID, "아이디txt")
    password = driver.find_element(By.ID, "비밀번호txt")
    username.send_keys("")  # 사용자 이름 입력
    password.send_keys("")  # 비밀번호 입력
    login_button = driver.find_element(By.ID, "확인imgbut")
    login_button.click()

    time.sleep(2)  # 페이지 로딩 대기
    # 페이지 이동
    driver.get("https://w3.a-b.kr/e5.aspx")

    # menu_link = driver.find_element(By.ID, "main_tab_id2")
    # menu_link.click()
    
    # 페이지 로딩 대기
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "DataGrid1"))
    )
    
    # 테이블 데이터 가져오기
    table = driver.find_element(By.ID, "DataGrid1")
    rowlen = len(table.find_elements(By.TAG_NAME, "tr"))
    
except Exception as e:
    print(f"로그인 오류 발생: {e}")
    driver.quit()

try:
    i = 1

    while i < rowlen:  # 첫 번째 행은 헤더이므로 제외
        table = driver.find_element(By.ID, "DataGrid1")
        rows = table.find_elements(By.TAG_NAME, "tr")
        row = rows[i]
        i += 1

        cols = row.find_elements(By.TAG_NAME, "td")
        if len(cols) > 0:
            name = cols[1].text.strip()  # 학생 이름
            checkbox = cols[0].find_element(By.CSS_SELECTOR, "input[type='checkbox']")
            
            if name in data:
                # 모든 체크 박스 해제
                for cb in driver.find_elements(By.CSS_SELECTOR, "input[type='checkbox']"):
                        if cb.is_selected():
                            cb.click()
                
                # 체크박스 선택
                checkbox.click()

                # 날짜 설정
                date_input = driver.find_element(By.ID, "날짜txt")
                date_input.click()
                date_input.clear()
                date_input.send_keys(TODAY)
                # 상담대상 선택 (값으로 선택)
                상담대상_select = Select(driver.find_element(By.ID, "상담대상combo"))
                상담대상_select.select_by_value("학생")

                # 교사 선택 (값으로 선택)
                교사_select = Select(driver.find_element(By.ID, "교사1combo"))
                교사_select.select_by_value("차진우")

                # textarea에 텍스트 입력
                textarea = driver.find_element(By.ID, "내용txt")
                textarea.click()
                textarea.clear()
                textarea.send_keys(data[name][0])  # 상담 내용 입력

                # 등록 버튼 클릭
                등록버튼 = driver.find_element(By.ID, "등록but")
                등록버튼.click()

                # Alert 확인
                alert = WebDriverWait(driver, 10).until(EC.alert_is_present())
                alert.accept()
                time.sleep(1)
except Exception as e:
    print(f"상담 등록 오류 발생: {e}")
    
    
    
```

```python

try:
    # 출결 페이지로 이동
    driver.get('https://w3.a-b.kr/o4.aspx')
    # 페이지 로딩 대기
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "DataGrid1"))
    )


    # 테이블 데이터 가져오기
    table = driver.find_element(By.ID, "DataGrid1")
    rowlen = len(table.find_elements(By.TAG_NAME, "tr"))

    i = 1
    while i < rowlen:  # 첫 번째 행은 헤더이므로 제외
        table = driver.find_element(By.ID, "DataGrid1")
        rows = table.find_elements(By.TAG_NAME, "tr")
        row = rows[i]
        i += 1
        cols = row.find_elements(By.TAG_NAME, "td")
        if len(cols) > 0:
            name = cols[1].text.strip()  # 학생 이름
            checkbox = cols[0].find_element(By.CSS_SELECTOR, "input[type='checkbox']")
            
            if name in data:
                # 모든 체크 박스 해제
                for cb in driver.find_elements(By.CSS_SELECTOR, "input[type='checkbox']"):
                        if cb.is_selected():
                            cb.click()
                
                # 체크박스 선택
                checkbox.click()

                # 날짜 설정
                date_input = driver.find_element(By.ID, "날짜txt")
                date_input.click()
                date_input.clear()
                date_input.send_keys(TODAY)

                # 상담대상 선택 (값으로 선택)
                상담자_txt = driver.find_element(By.ID, "상담자txt")
                상담자_txt.clear()
                상담자_txt.send_keys("차진우")

                # 교사 선택 (값으로 선택)
                구분_select = Select(driver.find_element(By.ID, "구분combo"))
                구분_select.select_by_value("출석")

                # textarea에 텍스트 입력
                textarea = driver.find_element(By.ID, "내용txt")
                textarea.click()
                textarea.clear()
                textarea.send_keys(data[name][1])  # 출결 내용 입력

                # 등록 버튼 클릭
                등록버튼 = driver.find_element(By.ID, "등록but")
                등록버튼.click()

                # Alert 확인
                alert = WebDriverWait(driver, 10).until(EC.alert_is_present())
                alert.accept()
                time.sleep(1)
except Exception as e:
    print(f"출결 등록 오류 발생: {e}")
    
finally:
    # 브라우저 종료
    driver.quit()
```
